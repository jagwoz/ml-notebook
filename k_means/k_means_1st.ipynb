```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.5)
```

```python
sns.scatterplot(x=data[0][:, 0], y=data[0][:, 1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters=3, n_init='auto')
k_means.fit(data[0])
```

```python
k_means.cluster_centers_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

ax1.set_title('KMeans')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=k_means.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')
```
