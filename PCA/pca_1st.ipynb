```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
data = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
```

```python
data.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(data)
```

```python
data_after_scale = scaler.transform(data)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(data_after_scale)
```

```python
pca_ = pca.transform(data_after_scale)
```

```python
print(data_after_scale.shape)
print(pca_.shape)
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=pca_[:, 0], y=pca_[:, 1], hue=dataset['target'], palette='rainbow')
plt.xlabel('1st Principal Component')
plt.ylabel('2nd Principal Component')
```

```python
pca.components_
```

```python
pca_comp_df = pd.DataFrame(pca.components_, columns=dataset['feature_names'])
```

```python
pca_comp_df
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(pca_comp_df, cmap='rainbow')
```
