```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
loans = pd.read_csv('data2.csv')
```

```python
loans.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=loans[loans['credit.policy'] == 1]['fico'], bins=30, color='blue')
sns.histplot(data=loans[loans['credit.policy'] == 0]['fico'], bins=30, color='red')
plt.legend(['Credit.Policy=1', 'Credit.Policy=0'])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=loans[loans['not.fully.paid'] == 1]['fico'], bins=30, color='blue')
sns.histplot(data=loans[loans['not.fully.paid'] == 0]['fico'], bins=30, color='red')
plt.legend(['Not.Fully.Paid=1', 'Not.Fully.Paid=0'])
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='purpose', hue='not.fully.paid', data=loans, palette='magma')
```

```python
plt.figure(figsize=(10, 6))
sns.jointplot(x='fico', y='int.rate', data=loans, color='red')
```

```python
plt.figure(figsize=(10, 6))
sns.lmplot(x='fico', y='int.rate', data=loans, hue='credit.policy', col='not.fully.paid', palette='Set1')
```
